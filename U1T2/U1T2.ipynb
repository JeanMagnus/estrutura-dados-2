```python
def maximum(values):
    # Inicializa a variável resposta como None
    answer = None
    # Percorre os valores na lista
    for value in values:
        # Se a resposta for None ou o valor atual for maior que a resposta
        if answer is None or value > answer:
            # Atribui o valor atual à resposta
            answer = value
    # Retorna a resposta que é o máximo encontrado na lista
    return answer

# Define a lista de valores de teste
test_values = [4, 3, 5, 6, 2, 1]
# Chama a função maximum() com os valores de teste e atribui o resultado a max_value
max_value = maximum(test_values)
# Imprime o valor máximo encontrado
print(max_value)
```

```python
def sum_pairs(values):
    pair_sums = 0              # c1
    for x in values:           # c2
        for y in values:       # c3
            pair_sums += x + y # c4
    return pair_sums           # c5

model1 = '(c3 + c4) * N^2 + c2 * N + (c1 + c5)'
model2 = 'c4 * N^2 + (c2 + c3) * N + (c1 + c5)'
model3 = '(c2 + c3 + c4) * N^2 + (c1 + c5)'

correct = model1
```



*   pair_sums = 0: Executada apenas uma vez, portanto, não depende de N.
*   for x in values: Executada N vezes, uma para cada elemento em values.
* for y in values: Executada N vezes para cada iteração do loop externo, resultando em N * N execuções.
* pair_sums += x + y: Executada N * N vezes, pois é executada para cada combinação de pares em values.
* return pair_sums: Executada apenas uma vez, portanto, não depende de N.

O modelo que representa melhor é o modelo 1.

```python
def sum_values(values):
    total = 0            # 1
    for value in values: # N
        total += value   # N
    return total         # 1

    # complexity = O(n)

def num_values(values):
    total = 0            # 1
    for _ in values:     # N
        total += 1       # N
    return total         # 1

    # complexity = O(n)

def average(values):
    value_sum = sum_values(values)  # N
    num_values = num_values(values) # N
    return value_sum / num_values   # 1

    # complexity = O(n)

```

```python
import matplotlib.pyplot as plt

def plot_costs(costs):
    plt.plot(costs, label='append cost')
    plt.ylabel('cost')
    plt.xlabel('N')
    plt.show()

def append_cost(array_length, list_length):
    if array_length == list_length:
        return array_length
    return 1

def append_N_list_cost(N):
    array_length = 1 # initially the array will have length 1
    list_length = 0 # initially the list has 0 elements
    total_cost = 0 # this variable will keep track of the total cost
    for i in range(N):
        cost = append_cost(array_length, list_length)
        total_cost += cost
        # update the array and list lengths
        if array_length == list_length:
            array_length *= 2
        list_length += 1
    return total_cost

# Inicialize uma lista vazia chamada custos
costs = []

# Use um loop for sobre range(5000) usando uma variável chamada N
for N in range(5000):
    # Chame custo_append_N_lista() passando o valor atual de N como argumento. Atribua o resultado a uma variável chamada custo.
    cost = append_N_list_cost(N)
    # Anexe o resultado à lista custos
    costs.append(cost)

# Use a função plot_costs() fornecida para plotar a lista de custos
plot_costs(costs)
```

```python
import time

def add_with_append(N):
    values = []
    for i in range(N):
        values.append(i)
    return values

def add_with_insert(N):
    values = []
    for i in range(N):
        values.insert(0, i)
    return values

# Medindo o tempo de execução para add_with_append(50000)
start = time.time()  # Capturando o tempo inicial
add_with_append(50000)
end = time.time()  # Capturando o tempo final
time_append = end - start  # Calculando o tempo de execução

# Medindo o tempo de execução para add_with_insert(50000)
start = time.time()  # Capturando o tempo inicial
add_with_insert(50000)
end = time.time()  # Capturando o tempo final
time_insert = end - start  # Calculando o tempo de execução

# Imprimindo os tempos de execução
print("Tempo de execução com append:", time_append)
print("Tempo de execução com insert:", time_insert)
```

```python
import random
import time
import matplotlib.pyplot as plt

def generate_list(k):
    items = []
    count = 1
    while len(items) < k:
        for _ in range(count):
            num1 = random.randint(10**(count-1), 10**count-1)
            num2 = random.randint(10**(count-1), 10**count-1)
            items.append((num1, num2))
            count += 1
            if len(items) == 5000:
                break
    return items

def plot_times(times):
    plt.plot(times)
    plt.ylabel('multiplication runtime')
    plt.xlabel('number of digits')
    plt.yscale('log')
    plt.yticks([])
    plt.show()

pairs = generate_list(5000)

# Inicialize uma lista vazia chamada times
times = []

# Use um loop for para iterar sobre todos os pares contidos na lista pairs
for x, y in pairs:
    # Obtenha o tempo antes da multiplicação
    start = time.time()
    # Multiplique x e y juntos
    _ = x * y
    # Obtenha o tempo depois da multiplicação
    end = time.time()
    # Adicione o tempo decorrido end - start à lista times
    times.append(end - start)

# Use a função plot_times() fornecida para plotar a lista times
plot_times(times)
```

```python
import random
import string
import time

def generate_list(k):
    result = []
    for _ in range(k):
        string_length = 10
        random_string = ''.join(random.choices(string.ascii_letters + string.digits, k=string_length))
        result.append(random_string)
    return result

def concat_with_add(word_list):
    concat = ''
    for word in word_list:
        concat += word
    return concat

def concat_with_join(word_list):
    return ''.join(word_list)

k = 100000  # Define o número de posições na lista
random_strings = generate_list(k)

# Medindo o tempo de execução para concat_with_add()
start = time.time()
concat_with_add(random_strings)
end = time.time()
time_add = end - start

# Medindo o tempo de execução para concat_with_join()
start = time.time()
concat_with_join(random_strings)
end = time.time()
time_join = end - start

# Imprimindo os tempos de execução
print("Tempo de execução com concat_with_add:", time_add)
print("Tempo de execução com concat_with_join:", time_join)
```
